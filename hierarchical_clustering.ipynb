# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_ah = cluster.fit_predict(X)
```

```python
print(y_ah)
```

## Visualising the clusters

```python
plt.scatter(X[y_ah==0,0], X[y_ah==0, 1], color='red', s=100, label='Cluster 1')
plt.scatter(X[y_ah==1,0], X[y_ah==1, 1], color='blue', s=100, label='Cluster 2')
plt.scatter(X[y_ah==2,0], X[y_ah==2, 1], color='green', s=100, label='Cluster 3')
plt.scatter(X[y_ah==3,0], X[y_ah==3, 1], color='orange', s=100, label='Cluster 4')
plt.scatter(X[y_ah==4,0], X[y_ah==4, 1], color='yellow', s=100, label='Cluster 5')
plt.title('Clusters of Customers (Hierarchical Clustering)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(bbox_to_anchor=(1.3, 1), loc="upper right")
plt.show()
```
